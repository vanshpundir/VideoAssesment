```python
from PyPDF2 import PdfReader
import re
import numpy as np
```

```python
def pdftotext(file):
    
    reader = PdfReader(file)

    # print(len(reader.pages))
    resumetext = ""
    for i in range(len(reader.pages)):
        page = reader.pages[i]
        text = page.extract_text()
        resumetext += text
    # print(text)
    return resumetext
```

```python
text1 = pdftotext("AIIntern.pdf")
text1 = re.sub("\\n", " ", text1)
```

```python
text2 = pdftotext("AssociateAIEngineer.pdf")
text2 = re.sub("\\n", " ", text2)
```
